```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("exam_score", axis=1)
y = df["exam_score"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
numeric_cols = ["age", "study_hours", "class_attendance", "sleep_hours"]
```

```python
ordinal_cols = ["sleep_quality", "facility_rating", "exam_difficulty"]
ordinal_categories = [["poor", "average", "good", "excellent"], #sleep_quality
                     ["low", "medium", "high"], #facility_rating
                     ["easy", "moderate", "hard"]] #exam_difficulty
```

```python
nominal_cols = ["gender", "course", "internet_access", "study_method"]
```

```python
preprocessor = ColumnTransformer(
    transformers=[("ordinal", OrdinalEncoder(categories=ordinal_categories), ordinal_cols),
                  ("onehot", OneHotEncoder(drop="first"), nominal_cols)],remainder="passthrough")
```

```python
X_train_encoded = preprocessor.fit_transform(X_train)
X_test_encoded = preprocessor.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
feature_names = preprocessor.get_feature_names_out()
print(feature_names)
```

```python
X_train_df = pd.DataFrame(X_train_scaled, columns=feature_names)
X_test_df = pd.DataFrame(X_test_scaled, columns=feature_names)
print(X_train_df.head())
```

```python
X_train_df.describe()
```

```python
X_train_df.columns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train_df, y_train)
```

```python
Y_predict  = model.predict(X_test_df)
```

```python
print(Y_predict)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print(r2_score(y_test, Y_predict))
```
